```python
import pandas as pd 
import numpy as np
import os 

def get_filename(number: int) -> str:
    try:
        for i in os.listdir("data"):
            keyword = str(number)+"_"
            if i[-3:] == 'csv' and i.startswith(keyword):
                filename = i # Only One!
        return filename
    except Exception:
        return 'NAN'

def data_processing(number: int) -> pd.DataFrame:
    '''
    para:
        number of subject
    '''
    k = get_filename(number)
    if get_filename(number) == 'NAN':
        return 'NAN'    
    else:
        filename = "data/"+ k 
    return pd.read_csv(filename)

raw_data = data_processing(1)
raw_data.head()
```

```python
labels = ['htsd','htnd','nthd','ntsd','sthd','stnd']
labels_explain = ['happy target and sad distraction','happy target and neutral distraction','neutral target and happy distraction','neutral target and sad distraction','sad target and happy distraction','sad target and neutral distraction']
rts = ['response_'+str(i+1)+'.rt' for i in range(6) ]
corrs = ['response_'+str(i+1)+'.corr' for i in range(6)]
tables = []
for item in range(len(labels)):
    # df.loc[:,['语文','数学']
    blocks = raw_data[raw_data.loc[:,'cond'] == labels[item]].loc[:,['number','corrAns','pleasantness','arousal','dominance','attraction',corrs[item],rts[item],'cond']]
    # blocks = raw_data[raw_data['cond'] == labels[item]].loc[:,['number','corrAns','pleasantness','arousal','dominance','attraction',corrs[item],rts[item],'cond']]
    tables.append(blocks)
# tables[0].loc[:,['corrAns','pleasantness','arousal','dominance','attraction','response_1.corr','response_1.rt','cond']]
tables[0]
```

```python
# report for the subject
def reporter_for_sub(blocks:list,filename:str)-> None:
    whole_trials = 12*6
    count = 0
    wh_rt = 0.0
    len_all = 0
    with open(filename,'w') as f:
        # f.write()
        for number in range(len(blocks)):
            f.write("Block: "+str(number+1)+"\n")
            f.write("Block information: "+labels_explain[number]+'\n')
            block_data = blocks[number]
            true_data = block_data[block_data[corrs[number]] == 1]
            corr_block_rate = len(true_data)/len(block_data)  # accurancy rate of one block
            f.write("Your accurancy rate is: "+str(corr_block_rate)+"\n")
            true_space_data = true_data[true_data['corrAns'] == 'space']# true and has space
            sum_block_rt = true_space_data[rts[number]].sum()
            avg_block_rt = sum_block_rt / len(true_data) # average reaction time in a block. reminder to cut error results"
            f.write("Your average reaction time is:"+str(avg_block_rt)+"\n\n")
            count += len(true_data)
            wh_rt += sum_block_rt
            len_all += len(true_space_data)
            print(len_all)
        f.write("\nWHOLE RESULTS:\n")
        f.write("Your whole accurancy rate is: "+str(count/72.0)+"\n")
        f.write("Your whole average reaction time is: "+str(wh_rt/len_all)+"\n")
    # pass
reporter_for_sub(tables,"1.txt")
```
